```python
#CRESCIMENTO DA POPULAO BRASILEIRA
import matplotlib.pyplot as plt



dados = open('populacao_brasileira.csv').readlines()

x = []
y = []

for i in range(len(dados)):
    if i != 0:
        linha = dados[i].split(';')
        x.append(int(linha[0]))
        y.append(int(linha[1]))
    

plt.bar(x,y, color='#e4e4e4')
plt.plot(x,y, color='k', linestyle = '--')
plt.title('CRESCIMENTO DA POPULAO BRASILEIRA')
plt.xlabel('Ano')
plt.ylabel('Populao x 100mm')
plt.show()
plt.savefig('Populao_brasileira.png')
```

```python
entrada = open('bacteria.fasta').read()
saida = open('bacteria.html', 'w')

cont = {}

for i in ['A', 'T', 'C', 'G']:
    for j in ['A', 'T', 'C', 'G']:
        cont[i + j] = 0

# Ajuste para considerar maisculas e minsculas
entrada = entrada.upper()

# Iterao sobre a entrada, considerando apenas as sequncias de nucleotdeos
for k in range(len(entrada)-1):
    if entrada[k] in 'ATCG' and entrada[k+1] in 'ATCG':
        cont[entrada[k] + entrada[k+1]] += 1

# HTML
saida.write('<div style="display: flex; flex-wrap: wrap;">')
i = 1

# Verifique se o valor mximo  diferente de zero
max_value = max(cont.values())
if max_value != 0:
    for k in cont:
        transparencia = cont[k] / max_value
        cor = f"rgba(0, 0, 255, {transparencia})"
        div_style = f"width: 100px; height: 100px; margin: 2px; background-color: {cor}; border: 1px solid #fff;"
        saida.write(f"<div style='{div_style}'></div>")
        
        if i%4 == 0:
            saida.write("<div> style= 'clear:both' ></div>")
        i+=1
else:
    print("Todos os valores so zero, no  possvel calcular a transparncia.")

saida.write('</div>')
saida.close()
```
